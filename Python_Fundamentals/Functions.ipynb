```python
def divideNumber(number):
    return number / 2
```

```python
divideNumber(20)
```

```python
myList = [3,4,5,6,7]
```

```python
dividingResultList = []
```

```python
for elm in myList:
    dividingResultList.append(divideNumber(elm))
```

```python
print(dividingResultList)
```

### map

```python
map(divideNumber,myList)
```

```python
list(map(divideNumber,myList))
```

```python
def strControl(string):
    return "a" in string
```

```python
strControl("adam")
```

```python
stringList = ["eylul","adam","anna"]
```

```python
list(map(strControl,stringList))
```

### filter

```python
list(filter(strControl,stringList))
```

### scope

```python
# L->Local, E-> Enclosing, G->Global, B->Built-In
```

```python
#Global
myString = "eylul global"

def myFunction():
    #Enclosing
    #myString = "eylul enclosing"
    print(myString)
    
    def myFunction2():
        #Local
        #myString = "eylul local"
        print(myString)
    
    myFunction2()
```
