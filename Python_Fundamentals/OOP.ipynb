```python
class Person:
    name = ""
    gender = ""
    age = 0
    job = "development"

    #initializer method (like constructer)
    # Assigns initial values to the object.
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender   # self references to itself (like 'this' in java)
        self.age = age

    #method
    def printName(self):    
        print(self.name)
```

```python
eylul = Person("Eylul Bencik","Female",21)
```

```python
eylul.printName()
```

```python
eylul.job = "Data Analyst"
print(eylul.job)
```

### inheritance

```python
animal_data = {
    "bird": "fly", 
    "road runner": ["fly", "run"]
}

class FlyingAnimal:
    
    def __init__(self, name):
        self.name = name
        
    def flying_info(self):
        
        info = animal_data.get(self.name)     # checks if the data contains the 'name'.
        if info and ("fly" in info or info == "fly"):
            print(f"{self.name.capitalize()} can fly.")
        else:
            print(f"{self.name.capitalize()} is not in our flying list.")

class Birds(FlyingAnimal):
    def __init__(self, name):
        super().__init__(name) # FlyingAnimal class's constructor.
                               # or use FlyingAnimal.__init__(self,name)

    def can_run(self):
        properties = animal_data.get(self.name)

        if isinstance(properties, list) and "run" in properties: # isinstance() checks if the 'properties' is a list.
            print(f"{self.name.capitalize()} can also run!")
        else:
            print(f"{self.name.capitalize()} cannot run.")

animal_name = input("Enter the name of animal (bird or road runner): ").lower().strip() # strip() removes spaces at the beginning and end

if animal_name in animal_data:
    my_bird = Birds(animal_name)
    my_bird.flying_info()
    my_bird.can_run()
    
#for precaution 
else:
    print(f"Error: '{animal_name.capitalize()}' not registered in our database.")
    print("Please just enter 'bird' or 'road runner'.")
```

### polymorphism

```python
class Company:

    def __init__(self,name):
        self.name=name

    def info(self):
        return f"Do you want to information about {self.name} ? "

class Staff:

    def __init__(self,name):
        self.name=name

    def info(self):
        return f"Do you want to information about {self.name} ? "

infoCompany = Company("ABC")
infoStaff = Staff("Umut")

infoCompany.info()
```

```python
infoStaff.info()
```

### encapsulation

```python
class UserEmail:

    def __init__(self,email,password):
        self.email  = email
        self.__password = password # __ works like private.(immutable)

    def info(self):
        print(f"{self.email}\nPassword is: {self.__password}\nYou cannot change this password.")
```

```python
myEmail = UserEmail("abc@mail.com",12345)
```

```python
myEmail.info()
```

```python
myEmail.password = 678
```

```python
myEmail.info()
```

### abstraction

```python
# reminder: from random import randint

from abc import ABC, abstractmethod
```

```python
class Car(ABC):

    @abstractmethod
    def maxSpeed(self):
        pass
```

```python
# you can't do myCar = Car()
```

```python
 # if you use an abstract class, your class must contain other class's properties.

class BMW(Car):
    def maxSpeed(self):
        print("It reaches speeds of 200 km per hour.")
    
myCar = BMW()    
```

```python
myCar.maxSpeed()
```

### Python Special Methods (Magic Methods)

```python
class Fruit:
    def __init__(self,name,calories):
        self.name = name
        self.calories = calories
        
# __str__ determines what will be displayed when the print(obj) command is executed.    
    def __str__(self):
        return f"{self.name}: {self.calories} calories."

myFruit = Fruit("pear",20)

print(myFruit)
```

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs  # songs is a list

# __len__ runs when the len(obj) function is called.
    def __len__(self):
        return len(self.songs)

my_hits = Playlist(["Song A", "Song B", "Song C"])
print(len(my_hits))
```

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __getitem__(self, index):
        return self.members[index]

my_team = Team(["Alice", "Bob", "Charlie"])
print(my_team[1]) 
```

```python
class Wallet:
    def __init__(self, balance):
        self.balance = balance

    def __add__(self, other):
        # Adds the balances of two Wallet objects
        return Wallet(self.balance + other.balance)

w1 = Wallet(50)
w2 = Wallet(30)
w3 = w1 + w2
print(w3.balance) 
```
