### args, kwargs (key word arguments)

```python
def args_sum(*args): # the user will decide how many arguments to use.
    return sum(args)
```

```python
args_sum(1,2,3,4,5,6)
```

```python
def kwargs_exmp(**kwargs):
    print(kwargs)
```

```python
kwargs_exmp(eylul=21,jake=22,paul=34)
```

```python
def kwargss(**kwargs):
    if "apple" in kwargs:
        print("there is apple")
    else:
        print("no")
```

```python
kwargss(apple=10,banana=43)
```

```python
def create_order_receipt(customer_name, *items, **special_requests):
    print(f"Order receipt for {customer_name}")
    
    # Using *args (comes as a Tuple)
    print("\nOrdered Items:")
    if not items:
        print("- No items selected.")
    else:
        for item in items:
            print(f"- {item}")
            
    # Using **kwargs {comes as a Dictionary}
    print("\nSpecial Requests and Notes:")
    if not special_requests:
        print("- No special requests.")
    else:
        for key, value in special_requests.items():
            print(f"{key}: {value}")
    print("-" * 35)
```

```python
# Scenario 1: Only the required argument and *args
create_order_receipt("Ahmet", "Hamburger", "French Fries", "Cola")
```

```python
# Scenario 2: Required argument, *args, and **kwargs all together
create_order_receipt("Eylul", "Sushi", "Noodles", napkin_count=5, sauce="Extra Soy", allergy="Peanuts")
```
