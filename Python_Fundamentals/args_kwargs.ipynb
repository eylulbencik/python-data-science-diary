### args, kwargs (key word arguments)

```python
def args_sum(*args): # the user will decide how many arguments to use.
    return sum(args)
```

```python
args_sum(1,2,3,4,5,6)
```

```python
def kwargs_exmp(**kwargs):
    print(kwargs)
```

```python
kwargs_exmp(eylul=21,jake=22,paul=34)
```

```python
def kwargss(**kwargs):
    if "apple" in kwargs:
        print("there is apple")
    else:
        print("no")
```

```python
kwargss(apple=10,banana=43)
```
