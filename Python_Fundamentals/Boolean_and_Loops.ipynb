### if-else

```python
age = input("Enter your age: ")
```

```python
if int(age) < 18:
    print("age < 18")
elif int(age) > 18 and int(age) <= 30:
    print("18 - age - 30")
elif int(age) > 30 and int(age) <= 40:
    print("30 - age - 40")
else:
    print("age > 40")
```

```python
listt = [1,2,3,40,50,60]
```

```python
for i in listt:
    print(i)
    if i == 40:
        print("yes")
        break
```
